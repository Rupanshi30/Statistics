## Covariance and correlation

### Covariance
Covariance is a metric that gives us the relationship between two random variables. This metric evaluates how much the variables change together. It is a measure of the linear relationship between two random variables.

Basically it tells if there is a relationship between two things and which direction that relationship is in.

It can take any positive and negative values, from -Ꝏ to +Ꝏ.

- Positive Covariance: It indicates that two variables tend to move in the same direction, which means that if we increase the value of one variable other variable value will also increase.
    
- Zero covariance: It indicates that there is no linear relationship between them.
    
- Negative covariance: It indicates that two variables tend to move in the opposite direction, which means that if we increase the value of one variable other variable value will decrease and vice-versa.
    
Covariance between two variables X and Y can be calculated :

    cov(x,y)= (xi-x)(yi-y)/n for population
    
    cov(x,y)= (xi-x)(yi-y)/n-1 for sample

### Limitations of Covariance

Covariance magnitude does not signify the strength of their relationship, so that only matters is the sign, whether it is positive or negative which tells the relationship.

Covariance is sensitive to scale, the covariance is measured in units. The unit/scale of the variables will be the unit of the covariance. And this is the reason, Covariance values are sensitive to the scale of the data and this makes them difficult to interpret. The value of the variance will differ if we change the scale/unit of the variable. 



## Correlation 
    - It is extracted from covariance 
    
    - Its formula is correlation = cov(x,y)/σx*σy
    


```python
import pandas as pd
l=[[650,19812,77200],[785,23926,998000],[1200,36576,1200000],[720,21945,800000],[975,29718,895000]]
df=pd.DataFrame(l,columns=["size_cm","size_ft","price"])
print(df)
cov=df.cov()  ##covariance
cov

```

- It depends on the unit, price is different for cm and ft

```python
 import pandas as pd
l=[[650,19812,77200],[785,23926,998000],[1200,36576,1200000],[720,21945,800000],[975,29718,895000]]
df=pd.DataFrame(l,columns=["size_cm","size_ft","price"])
print(df)
cor=df.corr()  ##correlation
cor

```

#### By default, correlation is Pearson

### Types of correlation

- **Pearson** : Mostly used for continuous variable
- **Spearman** : When we have **outliers in data** or you have **monotonic**, or **ordinal discrete data where rank** is associated

1. Pearson: Pearson correlation coefficient also referred to as Pearson's or the bivariate correlation is a statistic that measures the linear correlation between two variables X and Y.

2. Pearson correlation assumes that both variables are normally distributed.

3. It can be used for normal variables or continuous variables.

4. It has a value between +1 and -1. 
- A value of +1 is a total positive linear correlation
- 0 is no linear correlation 
- -1 is a total negative linear correlation

#### Note: The Pearson correlation can evaluate only a linear relationship between two.

#### Continuous Variables : A relationship is linear only when a change in one variable is associated with a proportional change in the other variable (change at constant rate)

### Limitation of Pearson Correlation:
1. It falls to capture the non linear relationship between two variables.

2. Usually, we do not use the Pearson correlation coefficient for ordinal variables(where sequence matters)

## 2. Spearman:
Spearman is a non parametric measure(no prior assumptions about distribution) of rank correlation(statistical dependence between the rankings of two variables).

    - It assesses how well the relationship between two variables can be described using a monotonic function.
    - The spearman correlation can evaluate a monotonic relationship between two variables - continuous or ordinal and it is based on the ranked values for each variable rather than the raw data.

### Probability Disturbution

```python
sample1 = [1,2,3,4,5,6] ## Discrete random variable for rolling a dice
df2 = pd.DataFrame(sample1,columns=['DiceRolls'])
df2
```

```python
prob=df2["DiceRolls"].value_counts(normalize=True)   ## normalize=Truein value_counts gives probability 
prob
```

```python
df2["Probability"]=prob.reset_index()["DiceRolls"]
df2
```

```python
import matplotlib.pyplot as plt
plt.bar(df2["DiceRolls"],df2["Probability"]);
```

```python
import numpy as np
cmf=np.cumsum(df2["Probability"])
cmf
```

```python
df2["cmf"]=cmf
df2
```

```python
plt.bar(df2["DiceRolls"],df2["cmf"]);
```

```python
iris=sns.load_dataset("iris")
iris["species"].unique()
```

```python
setosa=iris[iris["species"]=="setosa"]
versicolor=iris[iris["species"]=="versicolor"]
virginica=iris[iris["species"]=="virginica"]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(setosa["petal_length"])
```

```python
import numpy as np
np.histogram(setosa["petal_length"]) ## retuning two arrays one is count other is bins
```

```python
count_s,bins_s=np.histogram(setosa["petal_length"])
count_s,bins_s
```

```python
pdf_s=count_s/sum(count_s)
pdf_s
```

```python
plt.plot(bins_s[1:],pdf_s) ## have to start bins value from 1 so that index can be same (10-10)
```

- there is no species whose petal length is less than 1.1 and greater than 1.9

```python
cdf_s=np.cumsum(pdf_s)
cdf_s
```

```python
plt.plot(bins_s[1:],pdf_s) 
plt.plot(bins_s[1:],cdf_s) 
```

- 90 percentile setosa species whose petal length is less than 1.8
- cdf curve will attend 1 and stop there

- cdf play an important role to tell the particular performance of something with respect to others
- also tells the area less than the value
- when you do not have z-score use cdf

```python
counts,bins=np.histogram(virginica["petal_length"],bins=10)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bins[1:],pdf,label="virginica_pdf")
plt.plot(bins[1:],cdf,label="virginica_cdf")
counts,bins=np.histogram(versicolor["petal_length"],bins=10)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bins[1:],pdf,label="versicolor_pdf")
plt.plot(bins[1:],cdf,label="versicolor_cdf")
counts,bins=np.histogram(setosa["petal_length"],bins=10)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bins[1:],pdf,label="setosa_pdf")
plt.plot(bins[1:],cdf,label="setosa_cdf")
plt.legend()

```

- overlapping in versicolor and virginica but setosa is totally isolated
