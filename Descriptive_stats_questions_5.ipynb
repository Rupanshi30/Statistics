### find percentiles using z-score

#### Q1 suppose the scores on a certain college entrance exam are roughly normally distributed with a mean of 82 and a standard deviation of 5.If a certain student received a 90 on the exam. how well he performed in the exam relatively with respect to others. and also tell whether u r in top% or not


```python
## z = (x-mean)/std
z=(90-82)/5
z  ## left tail
```

This means that this student received a score that was 1.6 Standard deviations above the mean

- using z-table, 1.6=0.9452, dividing by 100 we get approx 95% that means student is in top 10%
